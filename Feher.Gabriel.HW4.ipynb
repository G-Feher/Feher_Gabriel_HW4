```python
import matplotlib.pyplot as plt
import numpy as np

mylabels = ['Andy', 'Amy', 'James', 'Jules', 'Arthur']
y = np.array([88,66,90,55,77])

plt.pie(y, labels = mylabels)
plt.legend(bbox_to_anchor=(1.0, 1.0))
```
